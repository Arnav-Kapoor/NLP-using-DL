```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import roc_auc_score, f1_score, confusion_matrix

from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
from tensorflow.keras.losses import BinaryCrossentropy
from tensorflow.keras.optimizers import Adam
```

```python
np.random.seed(0)
tf.random.set_seed(0)
```

```python
df=pd.read_csv('datasets/AirlineTweets.csv')
```

```python
df.head()
```

```python
df=df[['airline_sentiment','text']]
```

```python
df['airline_sentiment'].hist()
```

```python
df=df[df['airline_sentiment']!='neutral'].copy()
```

```python
target_map={'positive':1,'negative':0}
df['target']=df['airline_sentiment'].map(target_map)
```

```python
df.head()
```

```python
df_train,df_test=train_test_split(df,random_state=30)
```

```python
vectorizer=TfidfVectorizer(max_features=2000)
X_train=vectorizer.fit_transform(df_train['text'])
X_test=vectorizer.transform(df_test['text'])
```

```python
#tensorflow does not accept sparse matrix
X_train=X_train.toarray()
X_test=X_test.toarray()
```

```python
y_train=df_train['target']
y_test=df_test['target']
```

```python
#input dimensions
D=X_train.shape[1]
```

```python
#build model
input_layer=Input(shape=(D,))
output=Dense(1)(input_layer) #sigmoid included in loss function
model=Model(input_layer,output)
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.01),
             loss=BinaryCrossentropy(from_logits=True), #use from_logits=True when not applying sigmoid directly to dense layer
             metrics=['accuracy'])
```

```python
r=model.fit(X_train,y_train, validation_data=(X_test,y_test),epochs=40,batch_size=128)
```

```python
#plot loss per epoch
plt.plot(r.history['loss'],label='train loss')
plt.plot(r.history['val_loss'], label='val_loss')
plt.legend()
```

```python
#plot accuracy per epoch
plt.plot(r.history['accuracy'],label='train accuracy')
plt.plot(r.history['val_accuracy'],label='val_accuracy')
plt.legend()
```

```python
p_train=((model.predict(X_train)>0)*1.0).flatten()
p_test=((model.predict(X_test)>0)*1.0).flatten()
```

```python
cm=confusion_matrix(y_train,p_train,normalize='true')
cm
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['negative','positive'])
disp.plot()
```

```python
cm_test=confusion_matrix(y_test,p_test,normalize='true')
disp = ConfusionMatrixDisplay(confusion_matrix=cm_test, display_labels=['negative','positive'])
disp.plot()
```

```python
pred_train=model.predict(X_train)
pred_test=model.predict(X_test)
print('Train AUC score: ',roc_auc_score(y_train,pred_train))
print('Test AUC score: ',roc_auc_score(y_test,pred_test))
```

```python
print("Train F1: ",f1_score(y_train,p_train))
print("Test F1: ",f1_score(y_test,p_test))
```

```python
word_index_map=vectorizer.vocabulary_
word_index_map
```

```python
threshold=2

print("Most positive words:")
word_wight_tuples=[]
for word,index in word_index_map.items():
    weight=w[index,0]
    if weight>threshold:
        word_wight_tuples.append((word,weight))
word_wight_tuples=sorted(word_wight_tuples,key=lambda x:-x[1])
for i in range(10):
    word,weight=word_wight_tuples[i]
    print(word,weight)
```

```python
threshold=2

print("Most negative words:")
word_wight_tuples=[]
for word,index in word_index_map.items():
    weight=w[index,0]
    if weight<-threshold:
        word_wight_tuples.append((word,weight))
word_wight_tuples=sorted(word_wight_tuples,key=lambda x:x[1])
for i in range(10):
    word,weight=word_wight_tuples[i]
    print(word,weight)
```
